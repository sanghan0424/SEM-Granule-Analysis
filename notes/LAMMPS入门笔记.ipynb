## LAMMPS安装与启动
### 安装步骤
1. 下载预编译版本（Windows）→ 安装 → 配置环境变量（Path添加bin目录）
2. 验证：lmp -version 输出版本信息
### 基础脚本核心
1. create_box 2 box：创建2种原子类型（对应SEM亮/暗颗粒）
2. create_atoms：创建指定类型的原子
3. run 100：运行100步模拟
### 关联SEM
- 原子类型1 → thick_granule.jpg（亮颗粒）
- 原子类型2 → thin_granule.jpg（暗颗粒）
